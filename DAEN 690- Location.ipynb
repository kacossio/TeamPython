```python
#import 
import json
import tweepy as tw
import pandas as pd
import numpy as np
import datetime
import os
import time

#api keys
consumer_key = ''
consumer_secret = ''
access_key = ''
access_secret = ''

auth = tw.OAuthHandler(consumer_key, consumer_secret) 
auth.set_access_token(access_key, access_secret) 
api = tw.API(auth)
```

```python
def scraptweets(search_term, date_since, numTweets, numRuns):
    for i in range(0, numRuns):
        tweets = tw.Cursor(api.search, q=search_term, lang="en", since=date_since, tweet_mode='extended').items(numTweets)
        
    #varibles
        users_Vari = [[tweet.user.screen_name, 
               tweet.user.location,
               tweet.user.geo_enabled,
               tweet.user.description,
               tweet.user.profile_image_url[:63]+tweet.user.profile_image_url[70:]] for tweet in tweets]
    tweet_text = pd.DataFrame(data=users_Vari, 
                    columns=['user', "location", "geo_enable", "bio", "profile_image"])
    
    #drop the users'geo_enable is Fasle
    tweet_text.drop(tweet_text[tweet_text['geo_enable'] == False].index, inplace = True)
    #drop the location is empty
    tweet_text['location'].replace('', np.nan, inplace=True)
    tweet_text.dropna(subset=['location'], inplace=True)
    #drop the duplicate
    tweet_text.drop_duplicates(subset=['user'])
    from datetime import datetime
    timestamp = datetime.today().strftime('%Y%m%d_%H%M%S')
    time.sleep(920)
    path = os.getcwd()
    filename = path + '/project/' + timestamp + 'project.json'
    result = tweet_text.to_json(filename, index=False, orient="table")

#choose the search term and date since.
search_term = "e"+" -filter:retweets"
date_since = "2019-09-01"
#number of runs
numRuns = 2
#number of tweets for each run
numTweets = 1000
scraptweets(search_term, date_since, numTweets, numRuns)
```
